```python
from scipy.ndimage import filters
```

```python
plt.imshow(img)
```

```python
blurred_img = filters.gaussian_filter(img[100:200,100:200], 5)
```

```python
plt.imshow(blurred_img)
```

```python
blurred_img = np.zeros(img.shape)
```

```python
for i in range(3):
    blurred_img[:,:,i] = filters.gaussian_filter(img[:,:,i], 5)
```

```python
blurred_img = np.uint8(blurred_img)
```

```python
plt.imshow(blurred_img)
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_axis = np.arange(-10, 10, 0.001)
A = stats.norm.pdf(x_axis, 0, 2) #mean=0,SD=1
B = stats.norm.pdf(x_axis, 1, 2) #mean=1,SD=2
```

```python
plt.plot(x_axis, A);
plt.plot(x_axis, B);
plt.fill_between(x_axis, A, B, where=B>=A, facecolor='gray', interpolate=True);
plt.fill_between(x_axis, A, B, where=B<=A, facecolor='gray', interpolate=True);
```
